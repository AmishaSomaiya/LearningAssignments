import libraries

```python
import matplotlib.pyplot as plt
import numpy as np
from pylab import *
```

```python
%matplotlib notebook
```

Generate $rect$ signal as input and compute its $FFT$

```python
x = np.arange(-3, 3, 0.01)
y = np.zeros(len(x))
y[200:400] = 1
yShift = np.fft.fftshift(y)
fftyShift = np.fft.fft(yShift)
ffty = np.fft.fftshift(fftyShift)
```

Extract $magnitude$ and $phase$ data

```python
Xabs = abs(ffty)
Xphase = angle(ffty)
```

Run loops to generate plots of the input $rect$ signal, $magnitude$ plot and $phase$ plot depending upon value of iplot.

If iplot = 0, plot the subplots: input signal and $magnitude$ plot

If iplot = 1, plot the subplots: $magnitude$ plot and $phase$ plot

If iplot = 2, no plot

```python
iplot = 1

plt.rcParams['figure.figsize'] = (11,4)
plt.rcParams.update({'font.size':12})

if iplot == 0:
    subplot(2,1,1)
    plt.plot(y)
    plt.xlabel("Frequency (Hz)")
    plt.ylabel("Input Signal")
    
    subplot(2,1,2)
    plt.plot(Xabs)
    plt.xlabel("Frequency (Hz)")
    plt.ylabel("$|Magnitude|$")
    plt.subplots_adjust(hspace=0.4)
elif iplot == 1:
    subplot(2,1,1)
    plt.plot(Xabs)
    subplot(2,1,2)
    plt.plot(Xphase)
elif iplot == 2:
    print("No plot since iplot = 2")
else:
    print("Select iplot as 0, 1 or 2")
    

plt.show()
```
