# Part 1 Plotting a test signal

Plotting a test $sinc$ function

$$
f(x) =  \frac{\sin(x)}{x}
$$

```python
%config InlineBackend.figure_format = "svg"
pi = np.pi
x = np.linspace(-4*pi, 4*pi, 1000)
plt.rcParams['figure.figsize'] = (11,4)
plt.rcParams.update({'font.size':15})
plt.yticks([0,1], ['$0$', '$1$'])
plt.xticks([-4*pi, -2*pi, 0, 2*pi, 4*pi],
          ['$-4\pi$', '$-2\pi$', '$0$', '$2\pi$', '$4\pi$'])
plt.axhline(0, color='black',lw=1)
plt.plot(x, np.sin(x)/x, label=r'$f(x)=\frac{\sin(x)}{x}$')
plt.ylim(-0.5,1.5)
plt.title('The $sinc$ function')
plt.legend(loc='best', fontsize=20)
plt.show()
```

# Part 2 Compute and plot FFT of a periodic signal

Create a periodic test signal with sampling frequency f, base frequency f.

$$
x = 8sin(2pift) + 3sin(20pift)
$$

```python
f_s = 50.0 # Hz
f = 1.0 # Hz
time = np.arange(0.0, 3.0, 1/f_s)
x = 8 * np.sin(2 * np.pi * f * time) + 3 * np.sin(10 * 2 * np.pi * f * time)
```

Plot of signal x versus time

```python
plt.plot(time, x)
plt.xlabel("Time (sec)")
plt.ylabel("x")
```

Compute the Fast Fourier Transform of the input signal

```python
fft_x = np.fft.fft(x)
n = len(fft_x)
freq = np.fft.fftfreq(n, 1/f_s)
print (n)
print (freq)
```

```python
plt.plot(np.abs(fft_x))
```

Obtain 2 sided spectrum centred at zero by frequency shifting

```python
fft_x_shifted = np.fft.fftshift(fft_x)
freq_shifted = np.fft.fftshift(freq)

plt.plot(freq_shifted, np.abs(fft_x_shifted))
plt.xlabel("Frequency (Hz)")
```

Obtain plot for positive frequencies only. Normalize and obtain amplitude.  

```python
half_n = np.int(n/2.0)
fft_x_half = (2.0/n) * fft_x[:half_n]
freq_half = freq[:half_n]
```

```python
plt.plot(freq_half, np.abs(fft_x_half))
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
```

The magnitude of Fourier Cooefficients obtained in the above plot is the same as the amplitude of sinusoids of the test signal which verifies the code.
